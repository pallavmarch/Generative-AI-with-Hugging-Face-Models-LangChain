```python
from google.colab import userdata
sec_key = userdata.get("tokn")
os.environ["tokn"] = sec_key
```

```python
model_name = "deepset/roberta-base-squad2"

model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)
```

```python
QA_input = {
    'question': 'What is the only planet in our solar system to rotate clockwise on its axis?',
    'context': "Venus is the second planet from the Sun and is the only planet in our solar system to rotate clockwise on its axis. It is sometimes called Earth's 'sister planet' due to their similar size, mass, proximity to the Sun, and bulk composition. Venus has no moons and no rings"
}
res = nlp(QA_input)

print("Answer:", res["answer"])
print("Score:", res["score"])  # Confidence
```

```python
QA_input = {
    'question': 'What is the only planet in our solar system to rotate clockwise on its axis?',
    'context': "Venus is the second planet from the Sun and is the only planet in our solar system to rotate clockwise on its axis. It is sometimes called Earth's 'sister planet' due to their similar size, mass, proximity to the Sun, and bulk composition. Venus has no moons and no rings"
}

# Multiple answers
res = nlp(QA_input, top_k=3)

for i, answer in enumerate(res):
    print(f"Answer {i+1}: {answer['answer']}")
    print(f"Score: {answer['score'] * 100:.2f}% \n")
```
