```python
from google.colab import userdata
sec_key = userdata.get("tokn")
```

```python
os.environ["tokn"] = sec_key
```

```python
model_id = "gpt2"
model = AutoModelForCausalLM.from_pretrained(model_id)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
text_generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
result = text_generator("Venus is the only planet in our solar system that rotates clockwise on its axis", max_length=100, num_return_sequences=1, temperature=0.7)
print(result[0]['generated_text'])
```

```python
gpu_llm = HuggingFacePipeline.from_model_id(
    model_id="gpt2",
    task="text-generation",
    pipeline_kwargs={"max_new_tokens": 200, "num_return_sequences":1, "temperature":0.7}
)
```

```python
question = "What is the only planet in our solar system to rotate clockwise on its axis?"
template = """Question: {question} \n
Answer: Let's think step by step."""

prompt = PromptTemplate(template=template)

gpu_chain = LLMChain(llm=gpu_llm, prompt=prompt)


response = gpu_chain.run({"question": question})
print(response)
```
